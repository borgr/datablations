```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np

fig, axes = plt.subplots(nrows=7, ncols=1, figsize=(16, 12))

def frame_image(img, frame_width):
    b = frame_width # border size in pixel
    ny, nx = img.shape[0], img.shape[1] # resolution / number of pixels in x and y
    if img.ndim == 3: # rgb or rgba array
        framed_img = np.zeros((b+ny+b, b+nx+b, img.shape[2]))
    elif img.ndim == 2: # grayscale image
        framed_img = np.zeros((b+ny+b, b+nx+b))
    framed_img[b:-b, b:-b] = img
    return framed_img

def highlight_cell(x, y, w=1, h=1, ax=None, **kwargs):
    rect = plt.Rectangle((x-.5, y-.5), w, h, fill=False, **kwargs)
    ax = ax or plt.gca()
    ax.add_patch(rect)
    return rect

# https://stackoverflow.com/questions/18926031/how-to-extract-a-subset-of-a-colormap-as-a-new-colormap-in-matplotlib
def truncate_colormap(cmap, minval=0.0, maxval=1.0, n=100):
    new_cmap = colors.LinearSegmentedColormap.from_list(
        'trunc({n},{a:.2f},{b:.2f})'.format(n=cmap.name, a=minval, b=maxval),
        cmap(np.linspace(minval, maxval, n)))
    return new_cmap

cmap = truncate_colormap(plt.get_cmap('magma_r'), 0.2, 1.0)



LEN = 12

for ax, ep in zip(axes, range(1, 8, 1)):
    cur = 0
    step = LEN / ep
    vmin = 0
    vmax = 255
    vmin_selected = 0
    vmax_selected = 255 / ep

    for _ in range(ep):
        # https://stackoverflow.com/questions/24976471/matplotlib-rectangle-with-color-gradient-fill
        # https://matplotlib.org/2.0.2/examples/pylab_examples/gradient_bar.html
        ax.imshow(
          np.array([[vmin_selected, vmax_selected], [vmin_selected, vmax_selected]]),
          #cmap="gnuplot_r",
          #cmap="YlOrRd",
          #cmap="viridis_r",
          cmap=cmap,
          interpolation = 'bicubic', 
          vmin = vmin, vmax = vmax,
          extent=(cur, cur + step, 0, 1),
        )
        highlight_cell(0.5, 0.5, cur + step, 1, ax, color="black", linewidth=3)
        cur += step
    ax.set_xlim((0, LEN))
    txt = "EPOCH" if ep == 1 else "EPOCHS"
    ax.text(LEN//2, 0.5, f"{ep} {txt}", ha='center', va='center', color='black', weight='bold', fontsize=18)
    ax.axis('off')

plt.savefig("composition.pdf", bbox_inches='tight')
```
